```python
import pandas as pd

file_name = "ifn619_class_2026.csv"
class_df = pd.read_csv(f"data/{file_name}", index_col="id")
class_df.shape
```

```python
class_df.info()
```

```python
class_df.describe()
```

```python
class_df
```

```python
pd.set_option("display.max_rows", None)
class_df
```

```python
class_df.groupby("course").size()
```

```python
course_totals_df = class_df.groupby("course").size().reset_index(name="cohort_size")
course_totals_df
```

```python
course_totals_df.describe()
```

```python
import plotly.express as px

fig = px.bar(course_totals_df, x="course", y="cohort_size", width=600, height=400)
fig.show()
```

```python
sorted_df = course_totals_df.sort_values(by="cohort_size", ascending=False)
sorted_df
```

```python
fig = px.bar(sorted_df, x="course", y="cohort_size", width=600, height=400)
fig.show()
```

```python
class_df.groupby(["course", "mode"]).size()
```

```python
course_mode_df = class_df.groupby(["course", "mode"]).size().reset_index(name="course_mode")
course_mode_df
```

```python
fig2 = px.bar(course_mode_df, x="course", y="course_mode", color="mode", width=600, height=400)
fig2.update_xaxes(categoryorder="total ascending")
fig2.show()
```

```python
fig3 = px.bar(course_mode_df, x="mode", y="course_mode", color="course", width=600, height=400)
fig3.show()
```
